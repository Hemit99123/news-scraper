```python
from bs4 import BeautifulSoup
import requests

url = 'https://toronto.ctvnews.ca/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

news_items = soup.find_all('a', class_='c-list__item__link')[4:13]

top_news = []

for item in news_items:
    title = item.text
    link = item['href']

    top_news.append({'title': title, "link": link})

top_news
```
